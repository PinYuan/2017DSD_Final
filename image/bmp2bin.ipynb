```python
win_red = []
win_green = []
win_blue = []
for i in range(img_win.shape[0]):
    for j in range(img_win.shape[1]):
        win_red.append(img_win[i][j][0])
        win_green.append(img_win[i][j][1])
        win_blue.append(img_win[i][j][2])
```

```python
lose_red = []
lose_green = []
lose_blue = []
for i in range(img_lose.shape[0]):
    for j in range(img_lose.shape[1]):
        lose_red.append(img_lose[i][j][0])
        lose_green.append(img_lose[i][j][1])
        lose_blue.append(img_lose[i][j][2])
```

```python
tie_red = []
tie_green = []
tie_blue = []
for i in range(img_tie.shape[0]):
    for j in range(img_tie.shape[1]):
        tie_red.append(img_tie[i][j][0])
        tie_green.append(img_tie[i][j][1])
        tie_blue.append(img_tie[i][j][2])
```

```python
import pickle
rgb = []
rgb = win_red+win_green+win_blue+lose_red+lose_green+lose_blue+tie_red+tie_green+tie_blue
newFile = open("data.bin", "wb")
newFileByteArray = bytearray(rgb)
newFile.write(newFileByteArray)
```
