```python
from skimage import io
from skimage import transform
%matplotlib inline
```

```python
img_win=io.imread('win_resized.jpg')
io.imshow(img_win)
```

```python
img_lose=io.imread('lose_resized.jpg')
io.imshow(img_lose)
```

```python
img_win.shape
```

```python
img_win[0][0]
```

```python
win_red = []
win_green = []
win_blue = []
for i in range(img_win.shape[0]):
    for j in range(img_win.shape[1]):
        win_red.append(img_win[i][j][0])
        win_green.append(img_win[i][j][1])
        win_blue.append(img_win[i][j][2])
```

```python
lose_red = []
lose_green = []
lose_blue = []
for i in range(img_lose.shape[0]):
    for j in range(img_lose.shape[1]):
        lose_red.append(img_lose[i][j][0])
        lose_green.append(img_lose[i][j][1])
        lose_blue.append(img_lose[i][j][2])
```

```python
import pickle
rgb = []
rgb = win_red+win_green+win_blue+lose_red+lose_green+lose_blue
with open("data.bin", "wb") as mypicklefile:
    pickle.dump(rgb, mypicklefile)
```
